```python
import youtube_dl
import csv

# Fungsi untuk mengambil data dari YouTube
def scrape_youtube(keyword, start_date, end_date):
    # Opsi konfigurasi untuk youtube_dl
    ydl_opts = {
        'quiet': True,
        'no_warnings': True,
        'ignoreerrors': True,
        'skip_download': True,
        'outtmpl': '%(title)s.%(ext)s',
        'extract_flat': 'in_playlist',
        'daterange': youtube_dl.utils.DateRange(start_date, end_date),
        'noplaylist': True
    }

    # Kata kunci pencarian dan format tanggal
    search_keyword = f"ytsearchdateall:{keyword}"

    # Menggunakan youtube_dl untuk mencari dan mengambil data
    with youtube_dl.YoutubeDL(ydl_opts) as ydl:
        result = ydl.extract_info(search_keyword, download=False)

        # Menyimpan video dalam list untuk sorting
        videos = []
        if 'entries' in result:
            for video in result['entries']:
                if video is not None and (video.get('duration') is None or video.get('duration') > 60):
                    video['view_count'] = video.get('view_count') or 0
                    videos.append(video)

        # Mengurutkan video berdasarkan jumlah view (dari yang paling banyak)
        videos_sorted = sorted(videos, key=lambda x: x['view_count'], reverse=True)

        # Mencetak informasi video yang telah diurutkan
        for video in videos_sorted:
            print(f"Judul: {video.get('title', 'Tidak ada judul')}")
            print(f"Channel: {video.get('uploader', 'Tidak ada channel')}")
            print(f"Views: {video['view_count']}")
            print(f"Waktu: {video.get('upload_date', 'Tidak ada waktu')}")
            print(f"Link: https://www.youtube.com/watch?v={video.get('id')}")

            print("-" * 40)

    return videos_sorted
```

```python
# Fungsi untuk menyimpan data ke CSV
def save_to_csv(videos, filename):
    with open(filename, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(['Judul', 'Channel', 'Views', 'Waktu', 'Link'])  # Menulis header
        for video in videos:
            writer.writerow([
                video.get('title', 'Tidak ada judul'),
                video.get('uploader', 'Tidak ada channel'),
                video['view_count'],
                video.get('upload_date', 'Tidak ada waktu'),
                f"https://www.youtube.com/watch?v={video.get('id')}"
            ])

# Panggil fungsi dengan parameter yang diinginkan
videos_sorted = scrape_youtube("PILKADA", "20240801", "20240802")

# Simpan hasil scraping ke CSV
save_to_csv(videos_sorted, 'youtube_videos.csv')
```
